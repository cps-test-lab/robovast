# Update .vast Files - Change num_paths

## Configuration & Setup

Set `num_paths` value

```python
import os
import glob
import re

# User Configuration - Set your desired num_paths value here
TARGET_NUM_PATHS = 25  # Change this to your desired value

print(f"Target num_paths value: {TARGET_NUM_PATHS}")
```

## Find .vast Files

```python
# Define the dataset folder path
dataset_folder = "configs/dataset"

# Find all .vast files in the dataset directory
vast_files = glob.glob(os.path.join(dataset_folder, "*.vast"))

print(f"Found {len(vast_files)} .vast files:")
for file in vast_files:
    print(f"  - {os.path.basename(file)}")
```

```python
# Process all .vast files and update num_paths
num_paths_pattern = r'(\s*num_paths\s*:\s*)(\d+)'
modified_files = []

for file_path in vast_files:
    filename = os.path.basename(file_path)
    
    # Read file
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Check if file has num_paths
    if re.search(r'num_paths\s*:\s*\d+', content):
        # Replace any num_paths value with target value
        new_content = re.sub(num_paths_pattern, f'\\g<1>{TARGET_NUM_PATHS}', content)
        
        # Write back to file
        with open(file_path, 'w', encoding='utf-8') as f:
            f.write(new_content)
        
        modified_files.append(filename)
        print(f"✓ Updated {filename}")
    else:
        print(f"- {filename}: No num_paths found")

print(f"\nModified {len(modified_files)} files")
```
